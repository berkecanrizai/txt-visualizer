```python
!pip install wordcloud
!pip install fileupload
!pip install ipywidgets
!jupyter nbextension install --py --user fileupload
!jupyter nbextension enable --py fileupload
###some imports
import wordcloud
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import display
import fileupload
import io
import sys

###open file using utf8 encoding otherwise it can't handle some characters resulting in crash
file = open("2600-0.txt", encoding="utf-8")

file_contents = file.read()
```

```python
def calculate_frequencies(file_contents):
    #List of punctiations and some unwanted very common words. We will exclude them
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]
    count_words={}
    file_contents=file_contents.lower()
    splitted_words=file_contents.split()
    ##From the txt string, splitting each word and taking words as lower case
    for word in splitted_words:
        if word.isalpha() and (uninteresting_words.count(word)==0):
            if not word in count_words:
                count_words[word]=1
            else:
                count_words[word]=count_words[word]+1
    ## if word is alphabetical, meaning it doesn't have any numbers inside, and isn't part of boring words
    ## adding those words to dictionary where all occurences will be counted
    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(count_words)
    return cloud.to_array()
```

```python
### output as image
myimage = calculate_frequencies(file_contents)
plt.imshow(myimage, interpolation = 'nearest')
plt.axis('off')
plt.show()
a=4
```
